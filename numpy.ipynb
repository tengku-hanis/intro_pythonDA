Basics of numpy

In this notebook, we going to learn briefly about numpy.

**Import numpy**

```python
import numpy as np
```

**Create an array**

```python
arr = np.arange(6).reshape(3, 2)
arr
```

There are different ways to create an array

```python
# Create several arrays
arr2 = np.array([2, 3, 4])

arr3 = np.array([(2,3,4), (5,6,7)])

arr4 = np.zeros((2,3))

arr5 = np.ones((2,3))

# Print
print('This is arr2:\n', arr2, '\n')
print('This is arr3:\n', arr3, '\n')
print('This is arr4:\n', arr4, '\n')
print('This is arr5:\n', arr5, '\n')
```

**Types of an array**

```python
# 1d array
arr1d = np.arange(18)
arr1d
```

```python
# 2d array
arr2d = arr1d.reshape(6,3)
arr2d
```

```python
# 3d array
arr3d = arr1d.reshape(3,3,2)
arr3d
```

**Math operations**

```python
# Sum
np.sum(arr)
```

```python
# Mean
np.mean(arr)
```

```python
# Min max - this find the values
print(np.min(arr))
print(np.max(arr))
```

```python
# Argmin argmax - this find the index
print(np.argmin(arr))
print(np.argmax(arr))
```
