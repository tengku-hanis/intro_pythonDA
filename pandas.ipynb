**Create a new column**

```python
dat["ID"] = range(1, 769)
#id 769 is not inclusive
#numpy.arange(1, 779) is more efficient for large dataset

dat.head()
```

New column based on the old ones (categorical)


```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['Child', 'Young Adult', 'Middle Aged', 'Senior', 'Elderly']
dat['AgeGroup'] = pd.cut(dat['Age'], bins=bins, labels=labels)
dat.head()
```

```python
# Count the AgeGroup
dat['AgeGroup'].value_counts()
```

New column based on the old ones (numerical)


```python
dat['NewCol'] = dat['Pregnancies'] * 5
dat.loc[:, ['NewCol', 'Pregnancies']].head()
```
